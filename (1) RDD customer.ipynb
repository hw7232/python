```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf().setMaster('local').setAppName('customer')
```

```python
spark = SparkContext.getOrCreate(conf=conf)
```

## <span style="line-height:1.7">**customer-orders.csv**을 로드하여 <font color='blue'>**고객 ID를 key**로 하고, 고객이 지불한 **총 금액을 value**</font>로 하는 <b><font color='red'>RDD</font></b>를 찾으시오.</span>

### **customer-orders.csv**
- **comma**으로 구분 (',')
- <font color='blue'>**customer ID**, **order ID**, **payment**</font>
  > - <b>customer ID</b>: 고객 ID
  > - <b>order ID</b>: 주문 ID
  > - <b>payment</b>: 지불 금액

```python
customer_orders = spark.textFile("/content/customer-orders.csv")
```

```python
customer_orders_split = customer_orders.map(lambda x: x.split(","))
```

```python
customer_payments = customer_orders_split.map(lambda x: (x[0], float(x[2])))
```

```python
total_payments_by_customer = customer_payments.reduceByKey(lambda a, b: a + b)
```

```python
for customer, total_payment in total_payments_by_customer.collect():
    print(f"Customer ID: {customer}, Total Payment: {total_payment}")
```

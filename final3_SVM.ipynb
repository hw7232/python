**2.** **Diabetes.csv**파일을 로드하여, **linear SVM**을 학습하시오.
- Diabetes dataset은 혈당,BMI 등의 특징으로  당뇨를 판단하기 위한 dataset임
- 마지막 열인 **outcome은 y**, **나머지 열은 X**
- training data와 test data는 **7:3**으로 분리함 (반드시 random state를 지정한다.)
- SVM의 목표 함수인 $min \dfrac {1}{2} |w|^{2}+C \sum_{i} \xi_{i}$ 에서 $C$는 0.01로 설정함

```python
# prompt: 2. Diabetes.csv파일을 로드하여, linear SVM을 학습하시오.
# Diabetes dataset은 혈당,BMI 등의 특징으로 당뇨를 판단하기 위한 dataset임
# 마지막 열인 outcome은 y, 나머지 열은 X
# training data와 test data는 7:3으로 분리함 (반드시 random state를 지정한다.)
# SVM의 목표 함수인  min12|w|2+C∑iξi  에서  C 는 0.01로 설

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC

# Load the Diabetes dataset
diabetes = pd.read_csv('Diabetes.csv')

# Separate features and target
X = diabetes.iloc[:, :-1]
y = diabetes['Outcome']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Create a linear SVM classifier
clf = LinearSVC(C=0.01)

# Train the classifier
clf.fit(X_train, y_train)
```

```python
# prompt: 3. 위 2번에서 사용한 동일한 데이터를 사용하여 polynomial kernel SVM을 학습하시오.
# 2번 linear SVM과 성능을 비교할 수 있도록 반드시 동일한 training data와 test data를 사용한다.
# polynomial kernel의 degree는 3으로 지정하고, gamma와 coef0은 각각 1로 설정한다.
# Polynomial kernel:  (gamma×x⋅x′+coef0)degree=(x⋅x′+1)3

from sklearn.svm import SVC

# Create a polynomial kernel SVM classifier
poly_clf = SVC(kernel="poly", degree=3, gamma=1, coef0=1)

# Train the classifier
poly_clf.fit(X_train, y_train)

# Predict on the test set
poly_y_pred = poly_clf.predict(X_test)

# Compare the performance of the linear and polynomial kernel SVMs
from sklearn.metrics import accuracy_score

linear_accuracy = accuracy_score(y_test, y_pred)
poly_accuracy = accuracy_score(y_test, poly_y_pred)

print("Linear SVM accuracy:", linear_accuracy)
print("Polynomial SVM accuracy:", poly_accuracy)
```

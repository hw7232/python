```python
from pyspark import SparkConf, SparkContext
```

```python
spark_conf = SparkConf().setMaster('local').setAppName('hero')
spark = SparkContext.getOrCreate(conf=spark_conf)
```

## <span style="line-height:1.7">**Marvel-graph.txt**을 <b><font color='red'>RDD</font></b>로 로드하여 <b><font color='blue'>가장 많이 등장한 숫자</font></b>를 찾으시오.</span>
- **dict type**은 **spark.<font color='red'>parallelize</font>(dict)** 를 이용하여 **rdd로 변환**할 수 있다.

```python
marvel_graph = spark.textFile("Marvel-graph.txt")
```

```python
numbers = marvel_graph.flatMap(lambda line: line.split()).map(int)
```

```python
most_frequent_number = numbers.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b).max(lambda x: x[1])[0]
```

```python
print(f"Most frequent number: {most_frequent_number}")
```

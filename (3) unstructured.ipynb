```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
spark = SparkSession.builder.config('spark.driver.host','localhost').appName('unstructured').getOrCreate()
```

<font size='7'> <b>1.</b></font><font size='5'> <span style="line-height:1.7">아래와 같이 **learning.txt**의 <font color='blue'> **=== training process**</font> 다음 line부터 , <font color='blue'> **=== process termination**</font> 이전까지의 **각 line의 text를 string type의 값**으로 저장한 <font color='red'>**dataframe**</font>을 생성하시오.</span>
> **createDataFrame**을 이용하여 StringType의 값을 갖는 DataFrame을 생성할 수 있음

https://drive.google.com/file/d/1AwV9KH1pk6E9IyqG6UgIL78hMUvC-uOD/view?usp=drive_link

```python
df = spark.read.text("learning.txt")
```

```python
df = df.filter((col("value").like("%=== training process%")) | (col("value").like("%=== process termination%")))
```

```python
df = df.select(col("value").cast(StringType()))
```

```python
df.show(truncate=False)
```
